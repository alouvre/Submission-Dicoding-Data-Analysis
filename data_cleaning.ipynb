## Data Cleaning

```python
season = {
  1: "Springer",
  2: "Summer",
  3: "Fall",
  4: "Winter"
}

day_df["season"] = day_df["season"].map(season)
day_df.head()
```

```python
years = {
  0: "2011",
  1: "2012"
}

day_df["yr"] = day_df["yr"].map(years)
day_df.head()
```

```python
weathersit = {
  1: "Clear, Few clouds, Partly cloudy, Partly cloudy",
  2: "Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist",
  3: "Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds",
  4: "Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog"
}

day_df["weathersit"] = day_df["weathersit"].map(weathersit)
day_df.head()
```

```python
months = {
  1: "January",
  2: "February",
  3: "March",
  4: "April",
  5: "May",
  6: "June",
  7: "July",
  8: "August",
  9: "September",
  10: "October",
  11: "November",
  12: "December"
}

day_df["mnth"] = day_df["mnth"].map(months)
```

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

print(day_df.info())
print(hour_df.info())
```

```python
print(day_df.season.value_counts())
print(day_df.yr.value_counts())
print(day_df.weathersit.value_counts())
print(day_df.mnth.value_counts())
```

## Exploratory Data Analysis

```python
hour_df.describe(include="all")
```

```python
all_df = pd.merge(
  day_df,
  hour_df[['dteday', 'hr']],
  on='dteday',
  how='left'
)
```

```python
all_df.head()
```

```python
all_df.isna().sum()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x='hr', y='cnt', data=all_df)
plt.title("Hourly Distribution of Total Rentals")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='weathersit', y='cnt', data=all_df)
plt.title('Total Rentals for Different Weather Situations')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='season', y='cnt', data=all_df)
plt.title('Total Rentals for Different Seasons')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(all_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='holiday', y='cnt', data=all_df)
plt.title('Rentals on Holidays vs. Working Days')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='temp', y='cnt', data=all_df)
plt.title('Temperature vs Rentals')
plt.show()
```
